## Some preprocessing

```python
# I will split that raw data in rows and columns...
table = [row.split('\t') for row in data.split('\n')]

# and then create a dataframe
df = pd.DataFrame(table, columns=['Position', 'Bib', 'Name', 'AG', 'Time', 'M/F Position', 'Stared'])


# I'll convert the string-formated time into a float in minutes
def timeToMinutes(t):
    components = t.split(':')
    time =  float(components[0]) * 60 + \
            float(components[1]) + \
            float(components[2]) / 60
    return time

df['Minutes'] = df['Time'].apply(timeToMinutes)


# I'll extract the gender from the AG column
def getGender(s):
    if 'female' in s.lower():
        gender = 'female'
    elif 'male' in s.lower():
        gender = 'male'
    else:
        gender = 'unknown'
    return gender

df['Gender'] = df['AG'].apply(getGender)


# I'll extract the age group
def getAgeGroup(s):
    return ' '.join(s.split(' ')[1:])

df['Age'] = df['AG'].apply(getAgeGroup)


df.head()
```
